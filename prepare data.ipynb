```python

#----------------------------#
#    Import modules needed   #
#----------------------------#

import pandas as pd
from pandas import Series, DataFrame


#-----------------------------------------#
#    Create file paths for raw datasets   #
#-----------------------------------------#

path = r"C:\Users\stephanie.tuohey\OneDrive - Jisc\Steph\dashboards\Raw data"
crm = path + r"\report1643730004638.csv"
esfa = path + r"\College_accounts_academic_year_2019_to_2020_data_publication_v1.1.csv"
hesa = path + r"\dt051-table-1.csv"
hesafinance = path + r"\table-1.csv"

# print(crm)
# print(esfa)
# print(hesa)
# print(hesafinance)
```

```python
#----------------------------#
#    Create year variables   #
#----------------------------#

year = 2019

latest_year = str(year)+"/"+str(year+1)[-2:]
prev_year = str(year-1)+"/"+str(year)[-2:]
prev_year_2 = str(year-2)+"/"+str(year-1)[-2:]

print(latest_year)
print(prev_year)
print(prev_year_2)
```

```python
#-----------------------------------------------#
#    Import raw datasets as pandas dataframes   #
#-----------------------------------------------#

# keep relevant columns
crm_data = pd.read_csv(crm, engine = 'python').loc[:, ['Account Name', 
                                                       'Jisc Organisational ID',
                                                       'UKPRN',
                                                       'Jisc sector',
                                                       'Billing Country',
                                                       'Geographical Region']]

esfa_data = pd.read_csv(esfa, engine = 'python').loc[:, ['UKPRN',
#                                                          'College name',
                                                         'Total ESFA and Office for Students funded learners',
                                                         'Staff costs',
                                                         'Other operating expenses',
                                                         'Total Expenditure',
                                                         'Total staff_FTE_Number',
                                                         'Total Income',
                                                         'Other grants and contracts',
                                                         'Endowment and investment income',
                                                         'Funding body grants',
                                                         'Other income',
                                                         'Tuition fees and education contracts']]

hesa_data = pd.read_csv(hesa, engine = 'python',header=[17]).loc[:, ['UKPRN','Total']]

hesafin_data = pd.read_csv(hesafinance, engine = 'python',header=[12])
hesafin_data = hesafin_data.rename (columns={"Value(Â£000s)":"Value"})

list(hesafin_data)
# hesafin_data.head()
# esfa_data.head()
# hesa_data.head()
```

```python
#----------------------------------#
#    Filter HESA finance datasets  #
#----------------------------------#


he_income = hesafin_data[  (hesafin_data['Category'] =='Total income') 
                                & (hesafin_data["Region of HE provider"]=="All") 
                                & (hesafin_data["Country of HE provider"]=="All")
                                & (hesafin_data["Academic year"]==latest_year)
                                & (hesafin_data["Year End Month"]=="All")
                                ].rename (columns={"Value":"Income"})

he_expen = hesafin_data[  (hesafin_data['Category'] =='Total expenditure') 
                                & (hesafin_data["Region of HE provider"]=="All") 
                                & (hesafin_data["Country of HE provider"]=="All")
                                & (hesafin_data["Academic year"]==latest_year)
                                & (hesafin_data["Year End Month"]=="All")
                                ].rename (columns={"Value":"Expenditure"})
display(he_income)
display(he_expen)
```

```python
#----------------------------------#
#    Test CRM dataframe contents   #
#----------------------------------#

grouping_var = 'Jisc sector'
# grouping_var = 'Account Type'

crmdatatest=crm_data.groupby([grouping_var])["Account Name"].agg(['count'])
sorttest=crmdatatest.sort_values('count', ascending=False)
print(sorttest)
```

```python
#----------------------------#
#    Rename some variables   #
#----------------------------#
crm_df=crm_data.rename (columns={"Jisc Organisational ID": "JOID", 
                                 "Jisc sector": "Sector", 
                                 "Billing Country": "Country", 
                                 "Geographical Region": "Region"})
fec_df=esfa_data.rename (columns={"Total ESFA and Office for Students funded learners":"Learner headcount "+latest_year,
                                  "Total staff_FTE_Number": "Staff headcount "+latest_year})
hei_df=hesa_data.rename (columns={"Total":"Learner headcount "+latest_year})

list(crm_df)
list(fec_df)
list(hei_df)
```

```python
print(list(crm_df))
print(list(fec_df))
print(list(hesa_data))
print(list(he_income.loc[:, ["UKPRN","Income"]]))
```

```python
#--------------------------------------#
#    Merge FE data onto the CRM list   #
#--------------------------------------#

fe=crm_df.merge(fec_df,how='left',on='UKPRN')  # merging using UKPRN as a link

#--------------------------------------#
#    Merge HE data onto the CRM list   #
#--------------------------------------#

fe_he= fe.merge(hei_df,how='left',on='UKPRN')   # merging using UKPRN as a link

#---------------------------------------------#
#    Merge HE income data onto the CRM list   #
#---------------------------------------------#

fe_he_in= fe_he.merge(he_income.loc[:, ["UKPRN","Income"]],how='left',on='UKPRN')   # merging using UKPRN as a link

# display(fe_he_in)
# fe_he_in.head()
list(fe_he_in)
```
