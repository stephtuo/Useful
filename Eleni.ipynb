```python
# Import data
df = pd.read_excel('Leadership surveys 2021 - dashboard data for EDW.xlsx', header=[0, 1]) #full file
```

```python
q13= df[['ID.format','Q13a','Q13b','Q13c','Q13d','Q13e']] #isolating one question and the breakdowns I want ex. Sector, Role etc
q13 = q13.rename(columns={'ID.format':'Sector'})
q13.columns = q13.columns.droplevel(-1) #Removing multi level headers
q13.head()
```

```python
q13.groupby(['Sector']).describe()

# Percentage breakdown
print(q13.groupby(['Sector'])['Q13a'].value_counts(normalize=True) * 100 )

# counts breakdown
print(q13.groupby(['Sector'])['Q13a'].value_counts(normalize=False) )
```

```python
stats = pd.DataFrame()
stats['Count'] =  q13.groupby(['Sector'])['Q13a'].value_counts(normalize=False) #Adding both to dataframe
stats['Percentage'] = q13.groupby(['Sector'])['Q13a'].value_counts(normalize=True) * 100
stats #final dataframe can be stores as an excel and/or copied to a master excel
```
